```python
import pandas as pd
import numpy as np
```

```python
cleaned_df = pd.read_csv('data_cleaned.csv')
cleaned_df.head(2)
```

### Manual Tagging of Data

```python
#touchpoint and their relevant keywords

branch = ["slow", "wait", "branch", "counter", "customer", "counters", "friendly", "friendlier"]
call_center = ["answer", "call", "center", "centre", "ring", "rude", "friendly", "hotline", "careline", "feedback"]
social_media = ["facebook", "fb", "twitter", "insta", "instagram", "promote", "promotion"]
internet_banking = ["banking", "personnel", "account", "online", "internet", "login", "apply"]
mobile_banking = ["mobile", "app", "buggy", "laggy", "apps", "phone", "thumb","thumbprint"]
relationship_manager = ["rude", "friendly", "clueless", "manager", "staff", "loan", "rate", "interest"]
atm = ["spoil", "atm", "withdrawal", "deposit"]
```

```python
collective_tag = []

def common_elements(list1, list2):
    return list(set(list1).intersection(list2))


for i in cleaned_df["Verbatim Translated"]:
    individual_tag = []
    duplicated_removed = list(set(str(i).split()))
    if len(common_elements(branch,duplicated_removed)) > 0:
        individual_tag += ['branch']

    if len(common_elements(call_center,duplicated_removed)) > 0:
        individual_tag += ['call center']

    if len(common_elements(social_media,duplicated_removed)) > 0:
        individual_tag += ['social media']

    if len(common_elements(internet_banking,duplicated_removed)) > 0:
        individual_tag += ['internet banking']

    if len(common_elements(mobile_banking,duplicated_removed)) > 0:
        individual_tag += ['mobile banking']

    if len(common_elements(relationship_manager,duplicated_removed)) > 0:
        individual_tag += ['relationship manager']

    if len(common_elements(atm,duplicated_removed)) > 0:
        individual_tag += ['atm']

    individual_tag = ','.join([str(elem) for elem in individual_tag])

    if len(individual_tag) == 0:
        individual_tag = np.nan 

    collective_tag += [individual_tag]
    individual_tag = []


cleaned_df['Tagged Touchpoint'] = collective_tag
cleaned_df.dropna(inplace=True)
cleaned_df.head()
```

```python
cleaned_df.to_csv("tagged and feedback.csv")
```
